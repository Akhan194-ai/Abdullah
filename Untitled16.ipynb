```python
# Import necessary libraries
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Example dataset (replace with actual housing data)
data = {
    'Square_Feet': [1500, 2000, 2500, 1800, 2200, 1700, 1600, 3000, 1200, 2400],
    'Bedrooms': [3, 4, 4, 3, 4, 3, 2, 5, 2, 4],
    'Bathrooms': [2, 3, 3, 2, 3, 2, 1, 4, 1, 3],
    'Garage': [1, 2, 2, 1, 2, 1, 1, 3, 0, 2],
    'Price': [300000, 400000, 500000, 350000, 450000, 320000, 280000, 600000, 250000, 480000]  # Target variable
}

# Create a DataFrame
df = pd.DataFrame(data)

# Display the dataset
print("Dataset:")
print(df)

# Define features for clustering (excluding the target variable)
X = df[['Square_Feet', 'Bedrooms', 'Bathrooms', 'Garage']]

# Apply K-Means Clustering
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(X)

# Add cluster labels to the DataFrame
df['Cluster'] = kmeans.labels_

# Display the updated DataFrame
print("\nDataset with Cluster Labels:")
print(df)

# Visualize the clusters (using first two features for simplicity)
plt.figure(figsize=(8, 6))
plt.scatter(df['Square_Feet'], df['Bedrooms'], c=df['Cluster'], cmap='viridis', s=100)
plt.title('K-Means Clustering of Houses')
plt.xlabel('Square Feet')
plt.ylabel('Bedrooms')
plt.colorbar(label='Cluster')
plt.show()

# Example: Predict cluster for a new house
new_house = pd.DataFrame({
    'Square_Feet': [2000],
    'Bedrooms': [3],
    'Bathrooms': [2],
    'Garage': [1]
})
new_cluster = kmeans.predict(new_house)

print("\nPredicted Cluster for New House:")
print(new_cluster)
```
